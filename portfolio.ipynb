```python
import seaborn as sns # Seaborn

df_sample = df_cleaned.iloc[1:].sample(n=150, random_state=42) # Reduced number of data rows, which are selected randomly and reproducible
# This was done due to a long time taken from sns.pairplot() function
sns.pairplot(df_sample)
```

* Considering the plots, we see that GWP/POCP might show two clusters.
* Apart from this assumption, we can barely see two clusters - we also need to take into account that more than 5000 data rows are in this dataset.
* Hence, we set $k$ = 2.

```python
from sklearn.preprocessing import MinMaxScaler # MinMax Scaler
# from sklearn.preprocessing import StandardScaler # Standard Scaler

# Scaling the data
scaler = MinMaxScaler()
df_sc = pd.DataFrame(scaler.fit_transform(df_cleaned), columns=df_cleaned.columns)
df_sc.describe()
```

* The silhoutte score will be determined as well below.

```python
from yellowbrick.cluster import SilhouetteVisualizer # Silhouette Visualizer
from sklearn.metrics import silhouette_score # Silhoutte Score
from sklearn.cluster import KMeans # KMeans

silhouettes=[]

for k in range(2,11):
    kmeans=KMeans(n_clusters=k, random_state=1, n_init=10) # <- by default, this runs k-means++, reporting the best result
    cluster_assignments = kmeans.fit_predict(df_sc) 
    silhouettes.append(silhouette_score(df_sc,cluster_assignments))
```

```python
import matplotlib.pyplot as plt

plt.plot(range(2,11), silhouettes)
```

* Here we see that the silhoutte score is the highest at $k$ = 2.
* However, the silhoutte scores for $k$ > 2 are high as well.
* There is a small increase from $k$ = 3 to $k$ = 4.  
* For a good evaluation, we stick to $k$ = 2 since it has the highest score.

* Variance will also be calculated to compare the 4 features below.

```python
df_sc.var()
```

* Based on this, we can tell that the variances are close to each other.
* The values are also close to 0. The differences aren't so big and therefore, it is not possible to separate features based on this.

```python
import numpy as np

colors=['green', 'blue', 'darkmagenta']

def plot_clusters_with_centers(kmeans, scaler, X, title):
    '''
    create a 2 D plot for a dataset X, restricted to two features (index_x and index_y)
    use a given implementation of k-Means to cluster the data and plot clusters as colors and cluster centers (centroids) as stars.
    '''
    plt.figure(figsize=(8, 8))
    X_sc=scaler.fit_transform(X)
    cluster_assignments = kmeans.fit_predict(X_sc)

    cols=X.columns
    for i in range(0,len(np.unique(cluster_assignments))):
        plt.scatter(X[cluster_assignments==i][[cols[0]]], X[cluster_assignments==i][[cols[1]]], c=colors[i], label=i)

    cluster_centers=scaler.inverse_transform(kmeans.cluster_centers_)
        
    plt.scatter(
        cluster_centers[:, 0], 
        cluster_centers[:, 1],
        s=350, marker='*', c='red', edgecolor='black'
    )
    plt.title(title)
    plt.xlabel(cols[0])
    plt.ylabel(cols[1])
    plt.legend()
    plt.show()

    visualizer = SilhouetteVisualizer(kmeans, colors=colors, is_fitted=True)

    visualizer.fit(X_sc)        # Fit the data to the visualizer
    visualizer.finalize()
    print(f'Silhoutte score: {visualizer.silhouette_score_}')
    visualizer.show()        # Finalize and render the figure
    return cluster_assignments
```

### EM (Expectation-Maximization)

* Before performing EM algorithm on the DataFrame (GWP/POCP), we'll see if the data is in a Gaussian distribution.  

```python
import numpy as np

df_cleaned_gwp_pocp = np.random.normal(loc=0, scale=1, size=1000)

# Histogram with Gaussian fit
sns.histplot(df_cleaned_gwp_pocp, kde=True, stat='density', color='skyblue')
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Histogram with Gaussian Fit')
plt.show()
```

* Considering this histogram, we can see that it fits the Gaussian distribution more or less.  
* Now, we'll take steps to perform EM on this dataset.

```python
import matplotlib as mpl
from sklearn.mixture import GaussianMixture

# Ellipsis plot functions from the notebook of the lecture
def create_ellipse(gmm, ax, n, v0, v1, angle, color, alpha, stretch):
    ell = mpl.patches.Ellipse(
        gmm.means_[n, :2], stretch*v0, stretch*v1, angle=180 + angle, color=color
    )
    ell.set_clip_box(ax.bbox)
    ell.set_alpha(alpha)
    ax.add_artist(ell)


def make_ellipses(gmm, ax):
    for n, color in enumerate(colors):
        if gmm.covariance_type == "full":
            covariances = gmm.covariances_[n][:2, :2]
        elif gmm.covariance_type == "tied":
            covariances = gmm.covariances_[:2, :2]
        elif gmm.covariance_type == "diag":
            covariances = np.diag(gmm.covariances_[n][:2])
        elif gmm.covariance_type == "spherical":
            covariances = np.eye(gmm.means_.shape[1]) * gmm.covariances_[n]
        v, w = np.linalg.eigh(covariances)
        u = w[0] / np.linalg.norm(w[0])
        angle = np.arctan2(u[1], u[0])
        angle = 180 * angle / np.pi  # convert to degrees
        v = 2.0 * np.sqrt(2.0) * np.sqrt(v)

        create_ellipse(gmm, ax, n, v[0], v[1], angle, color, 0.5, 1)
        create_ellipse(gmm, ax, n, v[0], v[1], angle, color, 0.2, 2)
        create_ellipse(gmm, ax, n, v[0], v[1], angle, color, 0.1, 3)
```

```python
# Creating helper functions

# merge and shuffle clusters, create vector with the cluster_id (label)
def create_dataset(clusters):
    true_labels=[]
    for i in range(0, len(clusters)):
        true_labels.append(i*np.ones_like(clusters[i][:,0]))
    data=np.concatenate(clusters)
    labels=np.concatenate(true_labels)
    rng = np.random.default_rng()
    shuffeled_index=rng.permutation(len(data))
    return data[shuffeled_index], labels[shuffeled_index]
    

# run GaussianMixture Model and print resulting cluster properties    
def em_stats(data, n_clusters):
    gm=GaussianMixture(n_components=n_clusters)
    y_pred=gm.fit_predict(data)
    print("cluster sizes: ", np.bincount(y_pred))
    print("cluster means: \n", gm.means_)
    print("cluster covs: \n", gm.covariances_)
    return gm, y_pred

# plot clusters with data and distribution ellipses
def plot_clusters(clusters, colors, gm):
    fig=plt.figure(figsize=(8, 8))
    fig, ax = plt.subplots()
    for color, cluster in zip(colors, clusters):
        plt.scatter(cluster[:,0], cluster[:,1], c=color, s=2)
    if (gm): 
        make_ellipses(gm, ax)
    ax.set_aspect("equal", "datalim")
```
